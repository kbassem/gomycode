```
import matplotlib.pyplot as plt
import seaborn as sns

g=sns.FacetGrid(data_passengers,col='Sex')
g.map(plt.hist,'Age',bins=20)

grid = sns.FacetGrid(data_passengers,row='Pclass',col='Survived',size=2.2,aspect=1.6)
grid.map(sns.barplot,'Embarked','Age',alpha=.5,ci=None)
grid.add_legend()

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
    plot_correlation_map(data_passengers)
```
